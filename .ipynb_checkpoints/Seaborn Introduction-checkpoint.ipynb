```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planets = sns.load_dataset("planets")
```

```python
tips
```

```python
sns.scatterplot(x="tip", y="total_bill", data=tips, hue="day", size="size", palette="YlGnBu")
```

```python
sns.histplot(tips['tip'], kde=True, bins=15)
#kde=True, kde means kernel density estimate
#It is a smoothed, continuous version of the histogram, 
#which can help visualize the underlying distribution of the data.
```

```python
sns.displot(tips['tip'], kde=True, bins=15)
```

```python
sns.barplot(x="sex", y="tip", hue = "sex",  data = tips, palette = "YlGnBu")
#It is better to mention hue before palette
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue="sex", palette="YlGnBu")
```

```python
sns.stripplot(x="day", y="tip", data=tips, hue="sex", palette="YlGnBu")
```

```python
sns.jointplot(x="tip", y="total_bill", data=tips)
```

```python
sns.jointplot(x="tip", y="total_bill", data=tips, kind="reg")
```

```python
sns.jointplot(x="tip", y="total_bill", data=tips, kind="kde", fill=True)
```

```python
sns.jointplot(x="tip", y="total_bill", data=tips, kind="kde", fill=True, cmap = "YlGnBu")
```

```python
sns.jointplot(x="tip", y="total_bill", data=tips, kind="hex", cmap = "YlGnBu")
```

```python
sns.pairplot(titanic.select_dtypes(['number']), hue="pclass")
```

```python
# Select only numeric columns
numeric_titanic = titanic.select_dtypes(include=[float, int])
numeric_titanic.corr()

#because there is string datatype called 'male' in the titanic dataset 
#so it is good to mention that only float and int will be included in the corr matrix 
```

```python
sns.heatmap(numeric_titanic.corr(), annot=True, cmap="coolwarm")
```

```python
sns.clustermap(iris.drop("species", axis=1))
#because "species" is not a numeric datatype
#it can be useful for unsupervised learning
```
